```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#reading file
house_cost = pd.read_csv('./House_price.csv')
house_cost.describe()
```

```python
house_cost.head(5)
```

```python
house_cost.info()
```

```python
# area_income = house_cost['Avg. Area Income'].values
# house_price_values = house_cost['Price'].values
# plt.scatter(area_income,house_price_values)
# plt.show()
area_income = house_cost['Avg. Area Income'].values
house_price_values = house_cost['Price'].values
plt.scatter(area_income,house_price_values)
plt.show()
```

```python
x = area_income.reshape(-1,1)
x_train, x_test, y_train, y_test =train_test_split(x,house_price_values,train_size=0.8,test_size=0.2,random_state=100)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.xlabel('Av Area income')
plt.ylabel('Price Of House')
plt.title('Training data')
plt.show()
```

```python
linear = LinearRegression()
linear.fit(x_train,y_train)
y_predict = linear.predict(x_test)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_test,y_predict)
plt.xlabel("Average Area income")
plt.ylabel("Price of Houses")
plt.title("Trained model plot")
plt.plot
```

```python
Avg_income = 100000
price_predict = linear.predict([[Avg_income]])[0]
price_predict
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)*100

```

## 2- We can predicit the price of the house based on the average area income but is not accurate as we may have houses with low price and high income population.

## 3- It's hard to predict the house price depending only on the average income of the population as the area of the house highly affect its price.
